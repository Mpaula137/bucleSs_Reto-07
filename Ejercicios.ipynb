```python
def es_primo(numero):
    """
    Función que determina si un número es primo o no.
    Devuelve True si el número es primo, False si no lo es.
    """
    if numero <= 1:
        return False

    divisor = 2

    while divisor <= numero ** 0.5:
        if numero % divisor == 0:
            return False

        divisor += 1

    return True


numero = 1

while numero <= 100:
    if es_primo(numero):
        print(numero)

    numero += 1
```
